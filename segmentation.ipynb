```python
import os

opj = os.path.join

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def visualize(image, gts):
    '''
        image: np.asarray(PILLOW Image)
        gts: list of numpy gt
    '''
    result = np.copy(image).astype('uint8')
    for i, gt in enumerate(gts):
        result[gt!=0] = gt[gt!=0] /2 + image[gt!=0] /2
    result = Image.fromarray(result.astype('uint8'), 'RGB')
    plt.imshow(result)
```

```python
img_dir = '../../datasets/VOCdevkit/VOC2012/JPEGImages/2007_000032.jpg'
img = Image.open(img_dir)
imgnp = np.array(img).astype('uint8')
plt.imshow(img)
```

```python
gt_dir = '../../datasets/VOCdevkit/VOC2012/Binary_map_aug/train/1/2007_000032.png'
gt = Image.open(gt_dir).convert('RGB')
gtnp = np.array(gt).astype('uint8')
plt.imshow(gt)
```

```python
visualize(imgnp, [gtnp])
```
